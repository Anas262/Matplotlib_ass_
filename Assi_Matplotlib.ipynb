```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Qus:-2  What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# import numpy as np
# np.random.seed(3)
# x = 3 + np.random.normal(0, 2, 50)
# y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot

Ans:-2"""A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional plane. 
It is used to show the relationship between two variables and helps to identify patterns, clusters, outliers, and correlations in the data. """

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
colour=1000*np.random.rand(len(x))
size=500*np.random.rand(len(x))
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Random Data")
plt.scatter(x,y,c=colour,s=size,alpha=.8)
plt.show()
```

```python
# Qus:-3  Why is the subplot() function used? Draw four line plots using the subplot() function?
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans:-3""" The subplot() function is used to create multiple plots in a single figure, organized in a grid-like fashion. This is useful for 
comparing different plots or visualizing multiple aspects of your data side by side."""

x = np.array([0, 1, 2, 3, 4, 5]) 
#line 1:
y1= np.array([0, 100, 200, 300, 400, 500])
#line 2:
y2= np.array([50, 20, 40, 20, 60, 70])
#line 3:
y3=np.array([10, 20, 30, 40, 50, 60])
#line 4:
y4= np.array([200, 350, 250, 550, 450, 150])
#create 2*2 subplot
#first
plt.subplot(2,2,1)
plt.plot(x,y1,label="line 1")
plt.title("Line plot 1")
# Second
plt.subplot(2,2,2)
plt.plot(x,y2,label="line 2")
plt.title("Line plot 2")
#third
plt.subplot(2,2,3)
plt.plot(x,y3,label="line 3")
plt.title("Line plot 3")
#fourth
plt.subplot(2,2,4)
plt.plot(x,y4,label="line 4")
plt.title("Line plot 4")
#
plt.tight_layout()
plt.show()
```

```python
# Qus:-4 What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 
# import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
# profit = np.array([3000, 8000, 1000, 10000])

Ans:- """ A bar plot is a type of data visualization that represents data using rectangular bars, with the length or height of each bar 
proportional to the data value. It is used to compare values across different categories and is particularly useful for showing discrete data."""

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.title("Bar plot")
plt.bar(company,profit)
plt.show()
plt.ylabel("Company Name")
plt.xlabel("Profit")
plt.title("Bar plot")
plt.barh(company,profit)
plt.show()
```

```python
# Qus:-5 What is a box plot? Why is it used? Using the following data plot a box plot?

# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)
"""Ans:-5A box plot (also known as a whisker plot) is a type of data visualization that displays the distribution of a dataset along with 
important statistical information such as the median, quartiles,
and potential outliers. It is particularly useful for visualizing the spread and skewness of data."""

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2],labels=['Box 1','Box 2'])
plt.title("Box Plot")
plt.ylabel("value")
plt.show()
```
